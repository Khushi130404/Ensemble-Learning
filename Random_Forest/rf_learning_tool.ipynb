```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.tree import plot_tree,DecisionTreeClassifier
```

```python
x,y = make_classification(n_features=5,n_informative=5,n_clusters_per_class=1,n_redundant=0)
```

```python
df = pd.DataFrame(x,columns=['col1','col2','col3','col4','col5'])
df['target'] = y
print(df.shape)
df.head()
```

```python
def sample_rows(df,per):
    return df.sample(int(per*df.shape[0]),replace=True)
```

```python
def sample_cols(df,per):
    cols = random.sample(df.columns.tolist()[:-1],int(per*(df.shape[1]-1)))
    new_df = df[cols]
    new_df['target'] = df['target']
    return new_df
```

```python
def sample_combine(df,per_row,per_col):
    new_df = sample_rows(df,per_row)
    return sample_cols(new_df,per_col)
```

```python
df1 = sample_combine(df,0.5,0.5)
df2 = sample_combine(df,0.5,0.5)
df3 = sample_combine(df,0.5,0.5)
```

```python
clf1 = DecisionTreeClassifier()
clf2 = DecisionTreeClassifier()
clf3 = DecisionTreeClassifier()
```

```python
clf1.fit(df1.iloc[:,0:2],df1.iloc[:,-1])
clf2.fit(df2.iloc[:,0:2],df2.iloc[:,-1])
clf3.fit(df3.iloc[:,0:2],df3.iloc[:,-1])
```

```python
plot_tree(clf1)
```

```python
plot_tree(clf2)
```

```python
plot_tree(clf3)
```
